```python
age_ranges = range(0, 90, 5)
size_ranges = range(0, 100, 1)  
weight_ranges = range(0, 300, 20)  
height_ranges = range(0, 80, 5) 
rating_values = range(-1,10,1)

def count_fit_in_ranges(data, feature, ranges):
    range_counts = {r: {'small': 0, 'fit': 0, 'large': 0} for r in ranges}
    for entry in data:
        value = int(entry[feature])
        fit = entry['fit']
        for r in ranges:
            if r <= value < (r + (ranges[1] - ranges[0])):
                range_counts[r][fit] += 1
                break
    return range_counts


filtered_data_fit_ratings = [entry for entry in data if is_int(entry['rating']) and entry['rating'] in rating_values and entry['fit'] in ['small', 'large', 'fit']]

fit_counts_rating_values = count_fit_in_ranges(filtered_data_fit_ratings, 'rating', rating_values)
fit_counts_age_ranges = count_fit_in_ranges(filtered_data_fit_ages, 'age', age_ranges)
fit_counts_size_ranges = count_fit_in_ranges(filtered_data_fit_sizes, 'size', size_ranges)
fit_counts_weight_ranges = count_fit_in_ranges(filtered_data_fit_weights, 'weight', weight_ranges)
fit_counts_height_ranges = count_fit_in_ranges(filtered_data_fit_heights, 'height', height_ranges)

def plot_stacked_bar_chart(counts, ranges, title, xlabel):
    labels = [f"{r}-{r+(ranges[1] - ranges[0]) - 1}" for r in ranges]
    small_counts = [counts[r]['small'] for r in ranges]
    fit_counts = [counts[r]['fit'] for r in ranges]
    large_counts = [counts[r]['large'] for r in ranges]

    plt.figure(figsize=(10, 6))
    plt.bar(labels, small_counts, label='Small', color='blue')
    plt.bar(labels, fit_counts, bottom=small_counts, label='Fit', color='green')
    plt.bar(labels, large_counts, bottom=np.array(small_counts) + np.array(fit_counts), label='Large', color='red')
    plt.xlabel(xlabel)
    plt.ylabel('Count')
    plt.title(title)
    plt.legend()
    plt.show()

plot_stacked_bar_chart(fit_counts_age_ranges, age_ranges, 'Fit Distribution by Age Range', 'Age Range')
plot_stacked_bar_chart(fit_counts_size_ranges, size_ranges, 'Fit Distribution by Size Range', 'Size Range')
plot_stacked_bar_chart(fit_counts_weight_ranges, weight_ranges, 'Fit Distribution by Weight Range', 'Weight Range')
plot_stacked_bar_chart(fit_counts_height_ranges, height_ranges, 'Fit Distribution by Height Range', 'Height Range')
plot_stacked_bar_chart(fit_counts_rating_values, rating_values, 'Fit Distribution by Rating', 'Rating')
```
